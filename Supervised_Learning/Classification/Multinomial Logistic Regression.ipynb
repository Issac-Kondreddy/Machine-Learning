```python
# Importing necessary libraries
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Generating a synthetic dataset with 3 classes
X_multi, y_multi = make_classification(n_samples=1000, n_features=10, n_classes=3, n_informative=5, random_state=42)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Applying PCA to reduce the dataset to 2 dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_multi)
```

```python
# Plotting the first two principal components
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_multi, cmap='viridis', edgecolor='k', alpha=0.7)
plt.title('2D Visualization of the Dataset using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(*scatter.legend_elements(), title="Classes")
plt.show()
```

```python
# Splitting the dataset into a training set (70%) and a test set (30%)
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.3, random_state=42)
```

```python
# Creating and training a Multinomial Logistic Regression model
# 'multinomial' option specifies more than two classes; 'lbfgs' is the solver
multi_model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)
multi_model.fit(X_train_multi, y_train_multi)
```

```python
# Predicting labels for the test set and evaluating the model
y_pred_multi = multi_model.predict(X_test_multi)
accuracy_multi = accuracy_score(y_test_multi, y_pred_multi)
report_multi = classification_report(y_test_multi, y_pred_multi)

# Printing the accuracy and classification report
print("Accuracy:", accuracy_multi)
print("Classification Report:\n", report_multi)
```
